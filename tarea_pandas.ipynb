```python
winmag_url = 'https://gist.githubusercontent.com/clairehq/79acab35be50eaf1c383948ed3fd1129/raw/407a02139ae1e134992b90b4b2b8c329b3d73a6a/winemag-data-130k-v2.csv'
```

```python
df = pd.read_csv(winmag_url, index_col=0)
display(df)
```

```python
# cuantas variable tiene le dataframe?
num_variables = df.shape[1]
print(f"Tiene {num_variables} variables")
```

```python
# ¿Qué variables tienen valores perdidos?
valores_perdidos = df.isnull().sum()
print("Variables con valores perdidos:")
print(valores_perdidos[valores_perdidos > 0])
```

```python
# ¿Qué variables son numéricas?
variables_numericas = df.select_dtypes(include=['number']).columns
print("Variables numricas:")
print(variables_numericas)
```

```python
# ¿Qué variables son cualitativas?
variables_cualitativas = df.select_dtypes(include=['object']).columns
print("Las variables cualitativas:")
print(variables_cualitativas)
```

```python
# Hacer un dataframe con únicamente vinos europeos.
df_paises = df['country'].unique()
print(df_paises)
```

```python
# lista de países europeos
paises_europeos = [
    'France', 'Italy', 'Spain', 'Portugal', 'Germany', 'Austria',
    'Hungary', 'Greece', 'Switzerland', 'Belgium', 'Netherlands',
    'Czech Republic', 'Slovakia', 'Croatia', 'Serbia', 'Bulgaria',
    'Romania', 'Moldova', 'Luxembourg', 'Malta'
]
```

```python
#lista de paises europeos
df_europeos = df[df['country'].isin(paises_europeos)]
display(df_europeos)
```

```python
# ¿Cuál es el menor, el mayor y el precio promedio de la botella por país?
precios_por_pais = df.groupby('country')['price'].agg(['min', 'max', 'mean']).reset_index()

print("Precios por país (mínimo, máximo, promedio):")
print(precios_por_pais)


# ¿De que país es la botella de menor precio?

pais_menor_precio = df.loc[df['price'].idxmin()]['country']
print(f"El país de la botella de menor precio es: {pais_menor_precio}")
```
